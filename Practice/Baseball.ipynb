```python
import pandas as pd
url = 'http://npb.jp/bis/2018/stats/idp1_m.html'
dframe_list = pd.io.html.read_html(url)
dframe = dframe_list[0]
dframe
```

```python
dframe = dframe.drop(0)
dframe = dframe.drop(0,axis=1)
dframe.head()
```

```python
new_column = ['投手','登板','勝利','敗北','セ-ブ','ホ-ル','ＨＰ','完投','完封勝','無四球','勝率','打者','投球回','安打','本塁打','四球',
              '故意四','死球','三振','暴投','ボ-ク','失点','自責点','防御率']
```

```python
dframe.iloc[0,:]
```

```python
dframe = dframe.drop(14,axis=1)
```

```python
dframe = dframe.set_axis(new_column, axis='columns', inplace=False)
```

```python
dframe = dframe.drop(1)
```

```python
dframe.head()
```

```python
dframe.info()
```

```python
dframe["防御率"] = pd.to_numeric(dframe["防御率"])
```

```python
dframe["三振"] = pd.to_numeric(dframe["三振"])
```
